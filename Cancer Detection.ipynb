```python
X = df[['age','shape','margin','density']]
X
```

```python
X = df[['age','shape','margin','density']].values
X
```

```python
y = df[['severity']]
```

```python
y.head()
```

```python
y = df[['severity']].values
y
```

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler()
all_features_scaled = scaler.fit_transform(X)
all_features_scaled
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeClassifier(criterion = 'gini')
clf.fit(X_train,y_train)
```

```python
from IPython.display import Image  
from sklearn.externals.six import StringIO  
from sklearn import tree
from pydotplus import graph_from_dot_data 

dot_data = StringIO()  
tree.export_graphviz(clf, out_file=dot_data,  
                         )  
graph = graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())  
```

```python
clf.score(X,y)
```

Now applying the KFold technique

```python
from sklearn.model_selection import cross_val_score
clf = DecisionTreeClassifier()
cv_score = cross_val_score(clf , X,y,cv = 10)
cv_score.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 10,random_state = 1)
cv_score = cross_val_score(clf, X,y,cv = 10)
cv_score.mean()
```

```python
from sklearn.svm import SVC
from sklearn import svm
C = 1.0
SVC =  svm.SVC(kernel = 'rbf', C=C)
```

```python
cv_scores = cross_val_score(SVC, X,y,cv=10)
cv_score.mean()
```
